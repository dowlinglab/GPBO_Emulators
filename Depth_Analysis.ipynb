```python
import numpy as np
import pandas as pd
from bo_functions import get_sparse_grids
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 200
```

```python
#Pull Experimental data from CSV
n_exp = len(Xexp) 
n = 10
n_space = np.linspace(1,n,n)

plt.figure()
plt.xlabel("Dimensions (n)")
plt.ylabel("ln(len(weights))")
plt.title("Depth Analysis")
plt.grid(True)

depth = np.array([1,2,3,4,5])

for i in depth:
    log_len_weights = np.zeros(n)
    for j in n_space:
        weights = get_sparse_grids(int(j), output=0, depth = int(i))[1]      
        log_len_weights[int(j-1)] = np.log(len(weights))
    plt.plot(n_space,log_len_weights, label = "Depth ="+str(i))
plt.plot(n_space, np.full(n, np.log(1000), dtype=int), color = "black")
plt.plot([5,5], [0,np.amax(log_len_weights)], color = "black")
plt.legend(loc = "best")    
```
